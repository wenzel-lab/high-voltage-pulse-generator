```python
import tkinter as tk
from tkinter import messagebox

# Función para manejar el envío de datos de frecuencia
def set_frequency():
    try:
        frequency = int(frequency_entry.get())
        # Verifica que la frecuencia esté dentro del rango permitido
        if 50000 <= frequency <= 100000:
            print(f"Frequency set to: {frequency}")
            # Aquí puedes llamar a la función que actualiza la frecuencia en el hardware.
        else:
            messagebox.showerror("Error", "La frecuencia debe estar entre 50 y 100 kHz.")
    except ValueError:
        messagebox.showerror("Error", "Por favor, ingrese un valor numérico válido para la frecuencia.")

# Función para manejar el envío de datos de cada potenciómetro
def set_potentiometer_value(index):
    try:
        value = int(potentiometer_entries[index].get())
        if 0 <= value <= 255:  # Verifica que el valor esté en el rango permitido
            print(f"CH{index+1} position set to: {value}")
            # Aquí puedes llamar a la función que actualiza el valor del potenciómetro en el hardware.
        else:
            messagebox.showerror("Error", "El valor del potenciómetro debe estar entre 0 y 255.")
    except ValueError:
        messagebox.showerror("Error", "Por favor, ingrese un valor numérico válido para el potenciómetro.")

# Función para configurar todos los potenciómetros con un único valor
def set_all_potentiometers():
    try:
        value = int(all_potentiometer_entry.get())
        if 0 <= value <= 255:  # Verifica que el valor esté en el rango permitido
            for i, entry in enumerate(potentiometer_entries):
                entry.delete(0, tk.END)
                entry.insert(0, value)
            print(f"All potentiometer positions set to: {value}")
            # Aquí puedes llamar a la función que actualiza los valores del potenciómetro en el hardware.
        else:
            messagebox.showerror("Error", "El valor del potenciómetro debe estar entre 0 y 255.")
    except ValueError:
        messagebox.showerror("Error", "Por favor, ingrese un valor numérico válido para el potenciómetro.")

# Configuración de la ventana principal
root = tk.Tk()
root.title("Settings of HVPG")

# Etiqueta y campo de entrada para la frecuencia
tk.Label(root, text="Frequency (Hz):").grid(row=0, column=0, padx=10, pady=10)
frequency_entry = tk.Entry(root, width=10)  # Ancho ajustado
frequency_entry.insert(0, "80000")  # Valor por defecto de la frecuencia
frequency_entry.grid(row=0, column=1, padx=10, pady=10)

# Botón para enviar los datos de frecuencia
frequency_button = tk.Button(root, text="Set Value", command=set_frequency)
frequency_button.grid(row=0, column=2, padx=10, pady=10)

# Crear campos de entrada para los seis canales de potenciómetros
potentiometer_entries = []
for i in range(6):
    tk.Label(root, text=f"CH{i+1} position:").grid(row=i+1, column=0, padx=10, pady=5)
    entry = tk.Entry(root, width=10)  # Ancho ajustado
    entry.insert(0, "0")  # Valor por defecto de la posición
    entry.grid(row=i+1, column=1, padx=10, pady=5)
    potentiometer_entries.append(entry)
    
    # Botón para enviar los datos de cada potenciómetro
    button = tk.Button(root, text=f"Set CH{i+1}", command=lambda i=i: set_potentiometer_value(i))
    button.grid(row=i+1, column=2, padx=10, pady=5)

# Etiqueta y campo de entrada para la configuración de todos los potenciómetros
tk.Label(root, text="Set All Channels:").grid(row=7, column=0, padx=10, pady=10)
all_potentiometer_entry = tk.Entry(root, width=10)  # Ancho ajustado
all_potentiometer_entry.insert(0, "0")  # Valor por defecto de la posición
all_potentiometer_entry.grid(row=7, column=1, padx=10, pady=10)

# Botón para enviar los datos a todos los potenciómetros
all_potentiometers_button = tk.Button(root, text="Set All", command=set_all_potentiometers)
all_potentiometers_button.grid(row=7, column=2, padx=10, pady=10)

# Iniciar el bucle de la interfaz gráfica
root.mainloop()
```

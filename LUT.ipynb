```python
import cv2
import numpy as np

def apply_color_lut(image_path, output_path, color):
    # Cargar la imagen original
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        raise FileNotFoundError(f'Image not found at {image_path}')
    
    # Definir LUT para diferentes colores
    lut = np.zeros((256, 1, 3), dtype=np.uint8)
    
    if color == 'red':
        lut[:, 0, 2] = np.arange(256)  # Canal rojo
    elif color == 'green':
        lut[:, 0, 1] = np.arange(256)  # Canal verde
    elif color == 'blue':
        lut[:, 0, 0] = np.arange(256)  # Canal azul
    elif color == 'cyan':
        lut[:, 0, 0] = np.arange(256)  # Canal azul
        lut[:, 0, 1] = np.arange(256)  # Canal verde
    elif color == 'magenta':
        lut[:, 0, 0] = np.arange(256)  # Canal azul
        lut[:, 0, 2] = np.arange(256)  # Canal rojo
    elif color == 'yellow':
        lut[:, 0, 1] = np.arange(256)  # Canal verde
        lut[:, 0, 2] = np.arange(256)  # Canal rojo
    else:
        raise ValueError('Color not supported')
    
    # Aplicar el LUT
    colored_image = cv2.LUT(cv2.merge([image, image, image]), lut)
    
    # Guardar la imagen resultante
    cv2.imwrite(output_path, colored_image)
    print(f'Saved filtered image to {output_path}')

# Ejemplo de uso
image_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Video_3\Conteo\VmbImage_0012.bmp'
output_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Video_3\Conteo\VmbImage_0012_filtered_red.bmp'
apply_color_lut(image_path, output_path, 'yellow')  # Cambia 'red' por 'cyan', 'magenta', etc.
```

```python
import cv2
import numpy as np
from scipy.ndimage import label

def adjust_parameters(image):
    best_params = None
    max_count = 0
    
    # Intentar con diferentes combinaciones de parmetros
    for ksize in [3, 5, 7, 9]:
        for lower in range(30, 100, 10):
            for upper in range(150, 200, 10):
                blurred = cv2.GaussianBlur(image, (ksize, ksize), 0)
                edges = cv2.Canny(blurred, lower, upper)
                
                # En las versiones recientes de OpenCV, findContours devuelve solo 2 valores
                contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                
                # Contar contornos detectados
                count = len(contours)
                
                if count > max_count:
                    max_count = count
                    best_params = (ksize, lower, upper)
                    
    return best_params

def process_image(image_path, output_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Ajuste automtico de parmetros
    ksize, lower, upper = adjust_parameters(gray)
    print(f"Best parameters - GaussianBlur kernel: {ksize}, Canny thresholds: {lower}, {upper}")
    
    # Aplicar los mejores parmetros
    blurred = cv2.GaussianBlur(gray, (ksize, ksize), 0)
    edges = cv2.Canny(blurred, lower, upper)
    
    # Transformacin de distancia para separar nematodos
    dist_transform = cv2.distanceTransform(cv2.bitwise_not(edges), cv2.DIST_L2, 5)
    _, sure_fg = cv2.threshold(dist_transform, 0.5 * dist_transform.max(), 255, 0)
    sure_fg = np.uint8(sure_fg)
    unknown = cv2.subtract(cv2.bitwise_not(edges), sure_fg)
    
    # Marcadores para Watershed
    _, markers = cv2.connectedComponents(sure_fg)
    markers = markers + 1
    markers[unknown == 255] = 0
    
    markers = cv2.watershed(image, markers)
    image[markers == -1] = [0, 0, 255]  # Marcar bordes en rojo

    # Deteccin y etiquetado de nematodos
    contours, _ = cv2.findContours(sure_fg, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for i, contour in enumerate(contours):
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, f'Nematode {i+1}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
        cv2.putText(image, f'Pos: ({x},{y})', (x, y + h + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Guardar la imagen procesada
    cv2.imwrite(output_path, image)
    cv2.imshow("Processed Image", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Rutas de los archivos
image_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Test_O\ameba33pa14.png'
output_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Test_O\ameba33pa14.png'

# Procesar la imagen
process_image(image_path, output_path)

```

```python
import cv2
import numpy as np
import os
from datetime import datetime

def draw_circular_mask(image_shape, center, radius):
    """Dibuja una mscara circular en una imagen."""
    mask = np.zeros(image_shape[:2], dtype=np.uint8)
    cv2.circle(mask, center, radius, 255, thickness=cv2.FILLED)
    return mask

def define_regions(image_shape):
    height, width = image_shape[:2]

    # Define el centro de la imagen
    center_x, center_y = width // 2, height // 2

    # Define radios para las regiones circulares
    outside_grass_radius = int(min(width, height) / 2.0)  # Radio grande para fuera del csped
    inside_grass_radius = int(min(width, height) / 2.5)  # Radio pequeo para dentro del csped

    # Definir la mscara para el rea dentro del crculo exterior
    mask_inside_outside = draw_circular_mask(image_shape, (center_x, center_y), outside_grass_radius)

    # Definir la mscara para el rea dentro del crculo interior
    mask_inside = draw_circular_mask(image_shape, (center_x, center_y), inside_grass_radius)

    # Invertir la mscara para obtener el rea fuera del crculo interior pero dentro del crculo exterior
    mask_boundary = mask_inside_outside - mask_inside

    # Mscara para el rea fuera del crculo exterior
    mask_outside = np.zeros_like(mask_inside_outside)
    mask_outside[mask_inside_outside == 0] = 255

    return mask_outside, mask_boundary, mask_inside

def is_nematode(contour):
    """Determina si un contorno corresponde a un nematodo basado en su geometra."""
    # Calcular el rectngulo delimitador
    x, y, w, h = cv2.boundingRect(contour)
    
    # Relacin de aspecto: Proporcin entre ancho y alto
    aspect_ratio = float(w) / h
    
    # Calcular el rea y el rea del casco convexo
    area = cv2.contourArea(contour)
    hull_area = cv2.contourArea(cv2.convexHull(contour))
    
    # Solidez: Proporcin entre el rea del contorno y el rea del casco convexo
    solidity = float(area) / hull_area if hull_area != 0 else 0
    
    # Definir umbrales para identificar un nematodo
    if 0 < aspect_ratio < 1.5 and 0.5 < solidity < 0.95:
        return True
    return False

# Directorios
input_dir = r'C:\Users\Macarena\Pictures\videos_nematodos\Test_L'
output_dir = r'C:\Users\Macarena\Pictures\videos_nematodos\Test_O'

# Asegurarse de que el directorio de salida existe
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Procesar cada imagen en el directorio de entrada
for filename in os.listdir(input_dir):
    if filename.endswith(".png") or filename.endswith(".jpg"):  # Ajusta segn el tipo de imgenes
        # Cargar la imagen
        image_path = os.path.join(input_dir, filename)
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Aplicar un filtro de desenfoque para reducir el ruido
        blurred = cv2.GaussianBlur(gray, (7, 7), 0)

        # Deteccin de bordes utilizando Canny
        edges = cv2.Canny(blurred, 100, 120)

        # Dilatar los bordes para cerrarlos
        dilated = cv2.dilate(edges, None, iterations=2)

        # Encontrar los contornos
        contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Definir las regiones
        mask_outside, mask_boundary, mask_inside = define_regions(image.shape)

        # Inicializar contadores de nematodos
        nematodes_outside = 0
        nematodes_boundary = 0
        nematodes_inside = 0

        # Filtrar y etiquetar los nematodos
        nematode_count = 1
        for contour in contours:
            # Ignorar contornos muy pequeos
            if cv2.contourArea(contour) > 90 and is_nematode(contour):
                # Calcular el rectngulo delimitador
                x, y, w, h = cv2.boundingRect(contour)

                # Dibujar el rectngulo alrededor del nematodo
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

                # Etiquetar el nematodo
                label = f"Nematode {nematode_count} ({x}, {y})"
                cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 0, 0), 1)

                # Crear una mscara para el contorno actual
                contour_mask = np.zeros_like(mask_outside)
                cv2.drawContours(contour_mask, [contour], -1, 255, thickness=cv2.FILLED)

                # Verificar la interseccin con cada regin
                if np.any(cv2.bitwise_and(contour_mask, mask_outside)):
                    nematodes_outside += 1
                elif np.any(cv2.bitwise_and(contour_mask, mask_boundary)):
                    nematodes_boundary += 1
                elif np.any(cv2.bitwise_and(contour_mask, mask_inside)):
                    nematodes_inside += 1

                nematode_count += 1

        # Guardar la imagen etiquetada
        timestamp = datetime.now().strftime('%Y,%m,%d,%H,%M,%S')
        output_filename = f"{timestamp}.png"
        output_path = os.path.join(output_dir, output_filename)
        cv2.imwrite(output_path, image)

        # Mostrar la imagen etiquetada y los conteos de nematodos
        cv2.imshow('Labeled Nematodes', image)
        print(f"Imagen: {filename} - Nematodos en Outside: {nematodes_outside}, Boundary: {nematodes_boundary}, Inside: {nematodes_inside}")
        cv2.waitKey(0)
        cv2.destroyAllWindows()

print("Procesamiento completado.")
```

```python
import cv2
import numpy as np

def create_color_lut(color_name):
    """Crea un LUT para el color especificado."""
    lut = np.zeros((256, 1, 3), dtype=np.uint8)
    if color_name == 'red':
        lut[:, 0, 2] = np.arange(256)  # Rojo
    elif color_name == 'green':
        lut[:, 0, 1] = np.arange(256)  # verde
    elif color_name == 'blue':
        lut[:, 0, 0] = np.arange(256)  # azul
    elif color_name == 'cyan':
        lut[:, 0, 0] = np.arange(256)  # Rojo
        lut[:, 0, 1] = np.arange(256)  # Verde
    elif color_name == 'yellow':
        lut[:, 0, 1] = np.arange(256)  # Verde
        lut[:, 0, 2] = np.arange(256)  # Azul
    elif color_name == 'magenta':
        lut[:, 0, 0] = np.arange(256)  # Rojo
        lut[:, 0, 2] = np.arange(256)  # Azul
    else:
        raise ValueError(f"Color '{color_name}' no soportado")
    return lut

def draw_circular_mask(image_shape, center, radius):
    """Dibuja una mscara circular en una imagen."""
    mask = np.zeros(image_shape[:2], dtype=np.uint8)
    cv2.circle(mask, center, radius, 255, thickness=cv2.FILLED)
    return mask

def define_regions(image_shape):
    height, width = image_shape[:2]

    # Define el centro de la imagen
    center_x, center_y = width // 2, height // 2

    # Define radios para las regiones circulares
    outside_grass_radius = int(min(width, height) / 2.0)  # Radio grande para fuera del csped
    inside_grass_radius = int(min(width, height) / 2.5)  # Radio pequeo para dentro del csped

    # Definir la mscara para el rea dentro del crculo exterior
    mask_inside_outside = draw_circular_mask(image_shape, (center_x, center_y), outside_grass_radius)

    # Definir la mscara para el rea dentro del crculo interior
    mask_inside = draw_circular_mask(image_shape, (center_x, center_y), inside_grass_radius)

    # Invertir la mscara para obtener el rea fuera del crculo interior pero dentro del crculo exterior
    mask_boundary = mask_inside_outside - mask_inside

    # Mscara para el rea fuera del crculo exterior
    mask_outside = np.zeros_like(mask_inside_outside)
    mask_outside[mask_inside_outside == 0] = 255

    return mask_outside, mask_boundary, mask_inside

def apply_lut_to_region(image, lut, mask):
    """Aplica el LUT a una regin especfica usando una mscara."""
    print(f"Applying LUT. Image shape: {image.shape}, Mask shape: {mask.shape}")

    # Convertir la imagen a 3 canales si es en escala de grises
    if len(image.shape) == 2:
        image_3ch = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    else:
        image_3ch = image.copy()

    # Aplicar LUT a la imagen completa
    lut_applied = cv2.LUT(image_3ch, lut)
    print(f"LUT Applied. LUT shape: {lut.shape}")

    # Crear una imagen en blanco para aplicar la mscara
    result = np.zeros_like(image_3ch)

    # Aplicar la mscara al resultado LUT
    result[mask == 255] = lut_applied[mask == 255]
    print(f"Result Shape: {result.shape}")
    return result

def process_image(image_path, output_path):
    """Procesa la imagen aplicando LUT a las diferentes regiones."""
    # Leer la imagen en escala de grises
    gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    print(f"Gray Image Shape: {gray_image.shape}")

    # Definir las regiones
    mask_outside, mask_boundary, mask_inside = define_regions(gray_image.shape)

    # Crear LUTs para diferentes colores
    lut_red = create_color_lut('yellow')
    lut_yellow = create_color_lut('red')
    lut_cyan = create_color_lut('green')

    # Aplicar LUT a la regin exterior usando la mscara
    print("Applying LUT to Outside Region")
    colored_image = apply_lut_to_region(gray_image, lut_red, mask_outside)

    # Aplicar LUT a la regin del anillo usando la mscara
    print("Applying LUT to Boundary Region")
    boundary_image = apply_lut_to_region(gray_image, lut_yellow, mask_boundary)
    colored_image = cv2.add(colored_image, boundary_image)

    # Aplicar LUT a la regin interior usando la mscara
    print("Applying LUT to Inside Region")
    inside_image = apply_lut_to_region(gray_image, lut_cyan, mask_inside)
    colored_image = cv2.add(colored_image, inside_image)

    # Guardar la imagen resultante
    cv2.imwrite(output_path, colored_image)
    print(f"Output image saved to {output_path}")

# Ejemplo de uso
image_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Test_O\PA14+LT.png'
output_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Test_L\PA14+LT_lut.png'
process_image(image_path, output_path)

#codigo funciona correctamente, permite aplicar efecto LUT por regiones en imagenes 
```

```python
import xml.etree.ElementTree as ET

# Ruta del archivo XML
xml_file_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Resultofvideo_output6-1_Tracks.xml'

# Parsear el archivo XML
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Funcin para imprimir la estructura del XML
def print_xml_element(element, level=0):
    indent = '  ' * level
    print(f"{indent}{element.tag}: {element.attrib}")
    if element.text and element.text.strip():
        print(f"{indent}  Text: {element.text.strip()}")
    for child in element:
        print_xml_element(child, level + 1)

# Imprimir la estructura del XML
print_xml_element(root)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Ruta del archivo XML
xml_file_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Resultofvideo_output6-1_Tracks.xml'

# Leer el archivo XML
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Lista para almacenar los datos
data = []

# Iterar sobre cada particle
for particle in root.findall('particle'):
    nSpots = int(particle.get('nSpots'))
    for detection in particle.findall('detection'):
        t = detection.get('t')
        x = detection.get('x')
        y = detection.get('y')
        z = detection.get('z')
        data.append({
            'particle': len(data) // nSpots + 1,  # Identificador de la partcula
            't': t,
            'x': x,
            'y': y,
            'z': z
        })

# Crear DataFrame
df = pd.DataFrame(data)

# Guardar como CSV
csv_file_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Resultofvideo_output6-1_Tracks.csv'
df.to_csv(csv_file_path, index=False)

print(f'Datos guardados en {csv_file_path}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import LineCollection
from matplotlib.colors import Normalize
import matplotlib.cm as cm  # Uso de cm para obtener el colormap

# Ruta del archivo CSV
csv_file_path = r'C:\Users\Macarena\Pictures\videos_nematodos\Resultofvideo_output6-1_Tracks.csv'

# Leer el archivo CSV en un DataFrame
df = pd.read_csv(csv_file_path)

# Convertir las columnas a tipo float
df['t'] = df['t'].astype(float)
df['x'] = df['x'].astype(float)
df['y'] = df['y'].astype(float)
df['z'] = df['z'].astype(float)

# Seleccionar los datos de la primera partcula
df_first_particle = df[df['particle'] == 1]  # Asumiendo que '1' es la primera partcula

# Ordenar por tiempo para asegurar que la lnea siga el orden temporal
df_first_particle = df_first_particle.sort_values(by='t')

# Extraer datos para el grfico
points = df_first_particle[['x', 'y']].values
times = df_first_particle['t'].values
norm = Normalize(vmin=times.min(), vmax=times.max())
cmap = cm.get_cmap('viridis')  # Actualizado para evitar advertencia

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_xlim(points[:, 0].min() - 10, points[:, 0].max() + 10)
ax.set_ylim(points[:, 1].min() - 10, points[:, 1].max() + 10)
ax.set_xlabel('Posicin X')
ax.set_ylabel('Posicin Y')
ax.set_title('Trayectoria de la Primera Partcula con Degradado Temporal')

# Crear segmentos de lnea para el color degradado
segments = np.array([points[:-1], points[1:]]).transpose(1, 0, 2)
segment_times = (times[:-1] + times[1:]) / 2
colors = cmap(norm(segment_times))

# Crear la coleccin de lneas con colores degradados
lc = LineCollection(segments, colors=colors, linewidth=2)
ax.add_collection(lc)

# Aadir barra de color para representar el tiempo
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label('Tiempo')

# Mostrar el grfico
plt.show()
```

```python
import os
import re

# Directorio donde se encuentran las imgenes
image_folder = r'C:\Users\Macarena\Pictures\videos_nematodos\Video_3'

# Obtn una lista de todos los archivos en el directorio
files = os.listdir(image_folder)

# Filtra solo los archivos que empiezan con 'VmbImage_' y tienen la extensin '.bmp'
image_files = [f for f in files if f.startswith('Vmb_') and f.endswith('.bmp')]

# Funcin para extraer el nmero de un archivo
def extract_number(filename):
    match = re.search(r'(\d+)', filename)
    return int(match.group()) if match else 0

# Ordena los archivos por el nmero extrado
image_files.sort(key=extract_number)

# Renombra los archivos
for i, image_file in enumerate(image_files):
    # Extrae el nmero y renmbralo con ceros a la izquierda
    number = extract_number(image_file)
    new_name = f'Vmb_{number:04d}.bmp'
    old_path = os.path.join(image_folder, image_file)
    new_path = os.path.join(image_folder, new_name)

    # Renombra el archivo solo si el nuevo nombre es diferente y no existe
    if old_path != new_path and not os.path.exists(new_path):
        os.rename(old_path, new_path)
        print(f'Renombrado {old_path} a {new_path}')
    else:
        print(f'Error: No se puede renombrar {old_path} a {new_path} (puede que ya exista)')

print("Renombramiento completado.")
```

```python
import cv2

def resize_image(image, size=(800, 600)):
    height, width = image.shape[:2]
    new_size = size
    
    # Mantener la relacin de aspecto
    if width / height > new_size[0] / new_size[1]:
        scale = new_size[1] / height
        new_width = int(width * scale)
        new_height = new_size[1]
    else:
        scale = new_size[0] / width
        new_width = new_size[0]
        new_height = int(height * scale)
    
    resized_image = cv2.resize(image, (new_width, new_height))
    
    # Calcular el padding
    top = max((new_size[1] - new_height) // 2, 0)
    bottom = max(new_size[1] - new_height - top, 0)
    left = max((new_size[0] - new_width) // 2, 0)
    right = max(new_size[0] - new_width - left, 0)
    
    # Asegrate de que el tamao del padding sea vlido
    if top < 0 or bottom < 0 or left < 0 or right < 0:
        raise ValueError("Padding values should be non-negative.")
    
    # Agregar padding para ajustar el tamao exacto si es necesario
    padded_image = cv2.copyMakeBorder(resized_image, top, bottom, left, right, cv2.BORDER_CONSTANT, value=(0, 0, 0))
    
    return padded_image

# Uso
image = cv2.imread("C:/Users/Macarena/Pictures/videos_nematodos/Video_2/VmbImage_0002.bmp")
if image is not None:
    resized_image = resize_image(image, size=(800, 600))
    cv2.imwrite("resized_image.bmp", resized_image)  # Guarda la imagen redimensionada si es necesario
else:
    print("Error: Unable to read the image.")
```

```python
import cv2
import h5py
import numpy as np

def convert_avi_to_hdf5(avi_file, hdf5_file):
    # Abre el video AVI
    cap = cv2.VideoCapture(avi_file)
    
    # Obtiene el nmero de frames, ancho, y altura del video
    num_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    # Crea un archivo HDF5 para almacenar los datos del video
    with h5py.File(hdf5_file, 'w') as hdf:
        # Crea un grupo 'full_data'
        group = hdf.create_group('full_data')
        
        # Crea un dataset 'data' dentro del grupo 'full_data'
        dataset = group.create_dataset('data', shape=(num_frames, height, width), dtype=np.uint8)
        
        # Itera sobre los frames del video y los guarda en el dataset
        for i in range(num_frames):
            ret, frame = cap.read()
            if ret:
                # Convierte el frame a escala de grises
                gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                dataset[i] = gray_frame
            else:
                print(f"Error al leer el frame {i}")
                break
    
    # Libera el video
    cap.release()

# Ejemplo de uso
avi_file = r'C:\Users\Macarena\Pictures\videos_nematodos\video_output.avi'
hdf5_file = r'C:\Users\Macarena\Pictures\videos_nematodos\video_output.hdf5'
convert_avi_to_hdf5(avi_file, hdf5_file)
```
